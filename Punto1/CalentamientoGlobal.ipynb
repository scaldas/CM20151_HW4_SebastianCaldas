# Calentamiento Global?

### La idea de esta discusion es corroborar o cuestionar la nocion que se tiene que la actividad del hombre ha llevado a un aumento en la temperatura promedio de nuestra planeta. En este caso particular, se estudian datos para 5 ciudades Colombianas: Bogota,  Cali, Barranquilla, Bucaramanga e Ipiales. Se pretende usar herramientas estadisticas basicas para lograr lo anterior.  <br> <br> Se comienza entonces definiendo una funcion que, para cada ciudad, nos muestra tanto una interpolacion por splines de grado 1 como una regresion lineal hasta tercer orden, ambas extrapoladas al año 2017. Se espera entonces poder concluir a partir de esta informacion. No es necesario entender la funcion, solo correrla >^.^<

```python
import csv
from scipy.interpolate import interp1d, splrep, splev 

def print_interpolations(city):
    ff = open('temperaturas.csv', 'rt')
    reader = csv.reader(ff)

    dates = []
    hash_temperaturas = {}
    
    for row in reader:
        if row[4] == city and float(row[3]) != 999.9:
            date = datetime.datetime.strptime(row[5],"%Y-%m-%d")
            dates.append(date)
            hash_temperaturas[date] = float(row[3])  

    dates = np.sort(dates)

    temperaturas = []
    for each in dates:
        temperaturas.append(hash_temperaturas[each])

    dates = matplotlib.dates.date2num(dates)
    
    max_date = datetime.datetime.strptime('2017-12-31',"%Y-%m-%d")
    max_date = matplotlib.dates.date2num(max_date)
    
    x_l = np.linspace(dates[0],max_date,10000)
    tck = splrep(dates, temperaturas, k=1)
    y_spl = splev(x_l, tck)
    
    regression = polyfit(dates, temperaturas, 1) 
    y_reg = regression[0]*x_l + regression[1]
    
    figure(figsize=(8,6))
    plot_date(x_l, y_spl, '-b', c='k')
    plot_date(dates, temperaturas, 'o')
    title('Temperatura ' + city + ' - Interpolacion por splines')
    xlabel('Fecha')
    ylabel('Temperatura(ºC)')
    legend(['linear', 'data'])
    
    figure(figsize=(8,6))
    plot_date(x_l, y_reg, '-g', c='k')
    plot_date(dates, temperaturas, 'o')
    title('Temperatura ' + city + ' - Regresion lineal')
    xlabel('Fecha')
    ylabel('Temperatura(ºC)')
    legend(['linear', 'data'])
    
    print('Pendiente Regresion: ' + str(regression[0]))
```

## Bogota

```python
print_interpolations('Bogota')
```

### Arriba podemos observar el analisis para Bogota. La extrapolacion a partir de los splines es un metodo sesgado pues depende completamente del ultimo polinomio que se use para construirla. Mientras tanto, la regresion lineal que se hace muestra una clara tendencia al alza, con una pendiente positiva! 

## Barranquilla

```python
print_interpolations('Barranquilla')
```

### El comportamiento es el mismo que se describio para los datos de Bogota. Es importante anotar que la pendiente es bastante pequeña, pero es que se obtiene en las unidades en que maneja numpy las fechas. La tendencia creciente es clara.

## Cali

```python
print_interpolations('Cali')
```

### Vemos que la interpolacion por splines lleva a una extrapolacion que predice una caida en la temperatura de Cali para los proximos años. Esto es una evidencia que tal interpolacion es bastente dependiene de las condiciones de frontera al final del periodo. La pendiente de la regresion lineal sigue siendo positiva.

## Bucaramanga

```python
print_interpolations('Bucaramanga')
```

### El mismo caso que con Cali. La prediccion por splines predice una era del hielo para el 2017, pero ya habiamos concluido que esta prediccion es errada. La tendencia lineal parece apuntar a un crecimiento en las temperaturas.

## Ipiales

```python
print_interpolations('Ipiales')
```

### Finalmente, para la ciudad mas fria entre las estudiadas, encontramos la mayor pendiente para la regresion lineal. Las temperaturas parecen haber estado al alza (una vez mas).
